```python
import csv
import pandas as pd
import os
```

```python
def save_csv(dict_data, filename, folder_path, csv_columns):  
  # Check if the file exists
  if not os.path.isdir(folder_path):
    print("File create")
    os.mkdir(folder_path)
    
  try:
      with open(folder_path + filename, 'w') as csvfile:
          writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
          writer.writeheader()
          for data in dict_data:
              writer.writerow(data)
  except IOError:
      print("I/O error")
```

```python
df_process_month = df_process[~df_process["Months Code"].isin(drop_code)]
df_process_month = df_process_month.drop(["Element Code"], axis='columns')
df_process_month
```

```python
dict_year = {}

years = [ year for year in range(1996, 2022)]
month = df_process_month["Months"].unique().tolist()
countries = df_process_month["Area"].unique().tolist()

for y in years:

  dict_countries = {}

  for c in countries:
    dict_month = {}
    month_value = 1

    avg = 0.0
    for m in month:
      df_row_area = df_process_month[df_process_month["Area"] == c]
      df_row_month = df_row_area[df_row_area["Months"] == m]

      if len(df_row_month["Y{year}".format(year=y)].values) != 0:
        
        dict_month[month_value] =df_row_month["Y{year}".format(year=y)].values[0]
        avg = avg + df_row_month["Y{year}".format(year=y)].values[0]

      month_value = month_value + 1
    
    # Append the value to the countries
    dict_month["average"]  = avg / 12 
    dict_countries[c] = dict_month
 
  # Append to the year 
  dict_year[y] = dict_countries

dict_year
```
